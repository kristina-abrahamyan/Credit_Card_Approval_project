```python
import seaborn as sns
```

```python
from ipywidgets import interact
from ipyleaflet import Map
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
@interact(hue=['smoker','sex','time','day'])
def plot(hue):
    _=sns.pairplot(tips, hue=hue)
```
